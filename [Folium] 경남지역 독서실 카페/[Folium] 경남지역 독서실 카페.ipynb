```python
# -*- coding: utf-8 -*-

import pandas as pd
import numpy as np
import folium
```

```python
df = pd.read_csv(r"C:\Users\wlscj\Desktop\Data\소상공인데이터\소상공인시장진흥공단_상가(상권)정보_경남_202309.csv", low_memory=False,sep=',')
```

```python
study_cafe = df.loc[df.상권업종소분류명=='독서실/스터디 카페']
```

```python
study_cafe_Geoje = study_cafe.loc[study_cafe.시군구명=='거제시']
```

```python
study_cafe = study_cafe.reset_index()
study_cafe_Geoje = study_cafe_Geoje.reset_index()
```

```python
#경남 지역 대비 거제 지역 스터디 카페 수
Geoje_studycafe_number = study_cafe.shape[0]/ study_cafe_Geoje.shape[0]
print(Geoje_studycafe_number,'%')
```

```python
Gyeongnam_map = folium.Map(location=[35.5, 128.2132],zoom_start=9,titles='Gyeongnam_map')
```

```python
for i in range(len(study_cafe)):
    marker1 = folium.Circle([study_cafe.iloc[i]['위도'],study_cafe.iloc[i]['경도']],radius=200)
    marker1.add_to(Gyeongnam_map)
```
